```python
import pandas as pd

# Load your data
df = pd.read_csv("data/affairs.csv")

# Display first rows
print(df.head())

# Basic info
print(df.info())

# Summary statistics for numerical variables
print(df.describe())

# Check for missing values
print(df.isnull().sum())

# Frequency tables for categorical variables
print(df['occupation'].value_counts())
print(df['occupation_husb'].value_counts())

# Correlation matrix
print(df.corr())

# Average affair rate by religious level
print(df.groupby('religious')['affair'].mean())

# Average affair rate by years married groups
df['yrs_married_group'] = pd.cut(df['yrs_married'], bins=[0,5,10,20,30], right=False)
print(df.groupby('yrs_married_group')['affair'].mean())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Distribution of affair values
sns.histplot(df['affair'], bins=20)
plt.show()

# Relationship between age and affair rate
sns.scatterplot(x='age', y='affair', data=df)
plt.show()

# Boxplot: affair rate by religious level
sns.boxplot(x='religious', y='affair', data=df)
plt.show()
```
